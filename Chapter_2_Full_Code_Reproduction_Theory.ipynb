# What is Supervised Learning?

Supervised Learning adalah metode pembelajaran mesin yang menggunakan data berlabel (input dan target) untuk mempelajari hubungan antara keduanya.

# Classification vs Regression

Classification memprediksi kategori, sedangkan Regression memprediksi nilai numerik kontinu.

# Generalization

Tujuan utama machine learning adalah menghasilkan model yang bekerja baik pada data baru, bukan hanya data training.

# Overfitting

Model terlalu kompleks sehingga menghafal data training namun gagal pada data baru.

# Underfitting

Model terlalu sederhana sehingga gagal menangkap pola data.

## Sample Dataset: Breast Cancer Dataset

```

from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()

print(cancer.data.shape)
print(cancer.target_names)
```


### Penjelasan
Dataset Breast Cancer digunakan berulang kali dalam Chapter 2 untuk mengevaluasi berbagai algoritma supervised learning.


# k-Nearest Neighbors (KNN)

```

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

X_train, X_test, y_train, y_test = train_test_split(
    cancer.data,
    cancer.target,
    stratify=cancer.target,
    random_state=42
)

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

print("Train Accuracy:", knn.score(X_train, y_train))
print("Test Accuracy:", knn.score(X_test, y_test))
```


### Konsep Penting
KNN memprediksi kelas berdasarkan tetangga terdekat.
Parameter penting:
- n_neighbors
- metric
- weights


# Linear Models

```

from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(max_iter=5000)
logreg.fit(X_train, y_train)

print("Train:", logreg.score(X_train,y_train))
print("Test :", logreg.score(X_test,y_test))
```


### Penjelasan
Logistic Regression merupakan model linear untuk klasifikasi.
Walaupun namanya regression, model ini digunakan untuk classification.


# Decision Trees

```

from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(max_depth=4, random_state=0)
tree.fit(X_train,y_train)

print("Train:", tree.score(X_train,y_train))
print("Test :", tree.score(X_test,y_test))
```


Decision Tree membuat aturan keputusan berbentuk pohon.
Semakin dalam pohon, semakin tinggi risiko overfitting.


# Random Forest

```

from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(
    n_estimators=100,
    random_state=0
)

forest.fit(X_train,y_train)

print("Train:", forest.score(X_train,y_train))
print("Test :", forest.score(X_test,y_test))
```


Random Forest menggabungkan banyak Decision Tree untuk meningkatkan generalisasi.


# Gradient Boosting

```

from sklearn.ensemble import GradientBoostingClassifier

gbrt = GradientBoostingClassifier(random_state=0)
gbrt.fit(X_train,y_train)

print("Train:", gbrt.score(X_train,y_train))
print("Test :", gbrt.score(X_test,y_test))
```

# Support Vector Machine (SVM)

```

from sklearn.svm import SVC

svc = SVC()
svc.fit(X_train,y_train)

print("Train:", svc.score(X_train,y_train))
print("Test :", svc.score(X_test,y_test))
```


SVM mencari hyperplane terbaik untuk memisahkan kelas.
Konsep penting:
- Margin
- Support Vector
- Kernel Trick


# Neural Networks

```

from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(max_iter=1000, random_state=42)

mlp.fit(X_train,y_train)

print("Train:", mlp.score(X_train,y_train))
print("Test :", mlp.score(X_test,y_test))
```


### Penjelasan Neural Network

Komponen utama:
- Input Layer
- Hidden Layer
- Output Layer

Neural Network adalah fondasi Deep Learning modern.



# Perbandingan Algoritma

| Algoritma | Kelebihan | Kekurangan |

| KNN | Mudah dipahami | Lambat pada dataset besar |

| Decision Tree | Mudah dijelaskan | Overfitting |
| Random Forest | Akurasi tinggi | Lebih kompleks |
| SVM | Kuat pada data kecil-menengah | Mahal secara komputasi |
