# Code Reproduction 2: Sparse Matrix dengan SciPy

Buku menjelaskan bahwa data yang sebagian besar berisi nol lebih efisien disimpan menggunakan sparse matrix.


```

import numpy as np
from scipy import sparse

eye = np.eye(4)

sparse_matrix = sparse.csr_matrix(eye)

print(sparse_matrix)
```


### Penjelasan

Sparse Matrix menyimpan hanya elemen yang tidak bernilai nol sehingga menghemat memori.
Teknik ini sering digunakan pada:

# First Machine Learning Application

Pada Chapter 1, buku menggunakan Iris Dataset untuk membangun model klasifikasi pertama.


```

from sklearn.datasets import load_iris

iris_dataset = load_iris()

print(iris_dataset.keys())
```


### Penjelasan

Dataset Iris berisi:
- 150 sampel bunga iris
- 4 fitur
- 3 kelas

Kelas:
- Setosa
- Versicolor
- Virginica


```

print(iris_dataset['feature_names'])
print(iris_dataset['target_names'])
```


# Train-Test Split

Data harus dipisahkan menjadi:
- Training Set
- Testing Set

Tujuannya agar model diuji pada data yang belum pernah dilihat sebelumnya.


```

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'],
    iris_dataset['target'],
    random_state=0
)

print(X_train.shape)
print(X_test.shape)
```


### Penjelasan Penting

Chapter 1 menekankan bahwa menguji model pada data training adalah kesalahan.

Model harus dievaluasi menggunakan test set agar dapat mengukur kemampuan generalisasi.



# Building First Model: K-Nearest Neighbors


```

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=1)

knn.fit(X_train, y_train)
```


### Penjelasan Kode

`fit()` digunakan untuk melatih model.

Pada KNN:
- Tidak membangun persamaan matematis kompleks.
- Hanya menyimpan data training.
- Prediksi dilakukan dengan mencari tetangga terdekat.



# Making Predictions


```

import numpy as np

X_new = np.array([[5,2.9,1,0.2]])

prediction = knn.predict(X_new)

print(prediction)
print(iris_dataset['target_names'][prediction])
```


### Penjelasan

Model memprediksi spesies bunga berdasarkan fitur yang diberikan.



# Evaluating the Model


```

y_pred = knn.predict(X_test)

from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
```


### Penjelasan

Accuracy digunakan untuk menghitung proporsi prediksi yang benar.

Rumus:

Accuracy = Prediksi Benar / Total Data
