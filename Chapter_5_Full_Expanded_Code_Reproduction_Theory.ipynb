# Cross Validation

Cross Validation digunakan untuk memperoleh estimasi performa model yang lebih stabil dibanding train-test split biasa.

Ide utama:

Dataset dibagi menjadi beberapa fold.

Setiap fold bergantian menjadi test set.


```

from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

iris = load_iris()

scores = cross_val_score(
    LogisticRegression(max_iter=5000),
    iris.data,
    iris.target,
    cv=5
)

print(scores)
print(scores.mean())
```


### Penjelasan

cross_val_score()

Melatih dan mengevaluasi model pada beberapa fold.

cv=5

Menggunakan 5-Fold Cross Validation.



# Benefits of Cross Validation

Keuntungan:

1. Evaluasi lebih stabil
2. Memanfaatkan data lebih efisien
3. Mengurangi ketergantungan pada satu train-test split



# Stratified K-Fold Cross Validation

Pada klasifikasi, distribusi kelas perlu dipertahankan.

Stratified K-Fold menjaga proporsi kelas pada setiap fold.


```

from sklearn.model_selection import StratifiedKFold

kfold = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

print(kfold)
```


### Mengapa Penting?

Misalnya:

90% kelas A
10% kelas B

Jika tidak stratified, beberapa fold dapat kehilangan kelas minoritas.

# Simple Grid Search

Grid Search mencoba seluruh kombinasi hyperparameter.


```

from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

param_grid = {
    "n_neighbors":[1,3,5,7,9]
}

grid_search = GridSearchCV(
    KNeighborsClassifier(),
    param_grid,
    cv=5
)

grid_search.fit(
    iris.data,
    iris.target
)

print(grid_search.best_params_)
print(grid_search.best_score_)
```


### Penjelasan

GridSearchCV

Menguji seluruh kombinasi parameter.

Kemudian memilih kombinasi terbaik.



# The Danger of Overfitting the Validation Set

Jika terlalu sering mencoba parameter:

Model dapat overfit terhadap validation set.

# Metrics for Binary Classification

Confusion Matrix:

| | Pred Pos | Pred Neg |
|---|---|---|
| Actual Pos | TP | FN |
| Actual Neg | FP | TN |


```

from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    iris.data,
    iris.target,
    random_state=42
)

model = LogisticRegression(max_iter=5000)
model.fit(X_train,y_train)

pred = model.predict(X_test)

print(confusion_matrix(y_test,pred))
```


### Penjelasan

TP = True Positive

FP = False Positive

TN = True Negative

FN = False Negative



# Precision

Precision mengukur kualitas prediksi positif.


```

from sklearn.metrics import precision_score
```


Rumus:

Precision = TP / (TP + FP)



# Recall

Recall mengukur kemampuan menemukan seluruh kasus positif.



Rumus:

Recall = TP / (TP + FN)



# F1 Score

F1 merupakan harmonic mean antara precision dan recall.


```

from sklearn.metrics import classification_report

print(
    classification_report(
        y_test,
        pred
    )
)
```


### Penjelasan

classification_report()

Menampilkan:

- Precision
- Recall
- F1 Score
- Support
