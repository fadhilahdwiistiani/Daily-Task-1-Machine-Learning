Konsep terkenal:

# Categorical Variables

Variabel kategorikal berisi label atau kategori.

Contoh:
- Male / Female
- Jakarta / Bandung / Surabaya

Model Machine Learning tidak dapat langsung memproses teks sehingga perlu encoding.



# One-Hot Encoding

One-Hot Encoding mengubah kategori menjadi representasi numerik.

Contoh:

Color

Red
Blue
Green

Menjadi:

Red  Blue  Green

1      0      0
0      1      0
0      0      1


```

import pandas as pd

data = pd.DataFrame({
    "Color":["Red","Blue","Green","Blue"]
})

encoded = pd.get_dummies(data)

encoded
```


### Penjelasan Kode

pd.get_dummies()

Secara otomatis membuat kolom biner untuk setiap kategori.

Metode ini adalah dasar preprocessing data kategorikal.

# Binning and Discretization

Binning mengubah variabel kontinu menjadi interval.


```

import numpy as np

ages = np.array([18,25,31,45,60])

bins = np.digitize(
    ages,
    bins=[20,40,60]
)

print(bins)
```


### Penjelasan

Contoh:

18 → Kelompok 0

25 → Kelompok 1

45 → Kelompok 2

Binning sering digunakan pada:
- Risk Analysis
- Credit Scoring
- Customer Segmentation



# Linear Models vs Trees

Linear Models:

Mengasumsikan hubungan linear.

Decision Trees:

Mampu menangkap hubungan non-linear secara alami.

Karena itu feature engineering lebih penting pada model linear.



# Polynomial Features

Jika hubungan data tidak linear:

y = x²

Linear Regression biasa tidak cukup.

Solusi:

Polynomial Features


```

from sklearn.preprocessing import PolynomialFeatures

X = [[1],[2],[3],[4]]

poly = PolynomialFeatures(
    degree=2,
    include_bias=False
)

X_poly = poly.fit_transform(X)

print(X_poly)
```


### Penjelasan

Kolom baru:

x²

ditambahkan secara otomatis.

Model linear kemudian dapat mempelajari pola non-linear.



# Feature Interactions

Interaksi fitur terjadi ketika kombinasi dua fitur menghasilkan informasi baru.

Contoh:

Income × Education

Temperature × Humidity

PolynomialFeatures dapat membuat interaction features secara otomatis.


```

from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(
    degree=2,
    interaction_only=True,
    include_bias=False
)

X = [[1,2],[3,4]]

print(poly.fit_transform(X))
```

# Automatic Feature Selection

Dataset besar sering memiliki banyak fitur tidak relevan.

Feature Selection membantu memilih fitur yang paling informatif.



# Univariate Statistics

Memilih fitur berdasarkan hubungan statistik dengan target.


```

from sklearn.datasets import load_breast_cancer
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

cancer = load_breast_cancer()

selector = SelectKBest(
    score_func=f_classif,
    k=10
)

X_new = selector.fit_transform(
    cancer.data,
    cancer.target
)

print(X_new.shape)
```


### Penjelasan

SelectKBest

Memilih sejumlah fitur terbaik berdasarkan skor statistik.



# Model-Based Feature Selection

Menggunakan model Machine Learning untuk memilih fitur.


```

from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel

rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

rf.fit(cancer.data,cancer.target)

selector = SelectFromModel(
    rf,
    threshold="median"
)

X_selected = selector.transform(
    cancer.data
)

print(X_selected.shape)
```


### Penjelasan

Random Forest menghitung feature importance.

Fitur paling penting dipertahankan.



# Iterative Feature Selection

Metode memilih fitur secara bertahap.


```

from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=5000)

selector = RFE(
    model,
    n_features_to_select=10
)

X_rfe = selector.fit_transform(
    cancer.data,
    cancer.target
)

print(X_rfe.shape)
```


### Penjelasan

Recursive Feature Elimination (RFE)

Menghapus fitur yang kurang penting secara bertahap sampai jumlah fitur yang diinginkan tercapai.
