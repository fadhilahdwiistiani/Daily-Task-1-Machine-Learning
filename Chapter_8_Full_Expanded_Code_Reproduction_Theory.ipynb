# Scikit-Learn in Real Projects

Scikit-Learn sangat cocok untuk:

- Classification
- Regression
- Clustering
- Dimensionality Reduction
- Feature Engineering

Kelebihan:

- Mudah digunakan
- Dokumentasi lengkap
- Banyak algoritma siap pakai


```

from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression

iris = load_iris()

model = LogisticRegression(max_iter=5000)

model.fit(
    iris.data,
    iris.target
)

print(model.score(
    iris.data,
    iris.target
))
```


### Penjelasan

Contoh sederhana menunjukkan bagaimana Scikit-Learn dapat membangun model hanya dengan beberapa baris kode.



# Limitations of Scikit-Learn

Walaupun sangat kuat, Scikit-Learn memiliki keterbatasan:

- Deep Learning terbatas
- GPU support minimal
- Kurang cocok untuk data sangat besar
- Tidak dirancang untuk distributed training



# Neural Networks and Deep Learning

Deep Learning merupakan cabang Machine Learning yang menggunakan Neural Networks dengan banyak hidden layers.



## Komponen Neural Network

- Input Layer
- Hidden Layer
- Output Layer
- Weights
- Activation Functions


```

from sklearn.neural_network import MLPClassifier
from sklearn.datasets import load_digits

digits = load_digits()

mlp = MLPClassifier(
    hidden_layer_sizes=(100,),
    max_iter=1000,
    random_state=42
)

mlp.fit(
    digits.data,
    digits.target
)

print(
    mlp.score(
        digits.data,
        digits.target
    )
)
```


### Penjelasan

MLPClassifier merupakan implementasi Neural Network sederhana di Scikit-Learn.

Untuk Deep Learning modern biasanya digunakan:

- TensorFlow
- Keras
- PyTorch
