# Why Pipelines Matter

Pada proyek Machine Learning nyata, workflow biasanya terdiri dari:

Data
↓
Preprocessing
↓
Feature Engineering
↓
Model Training
↓
Prediction

Pipeline membantu mengotomatisasi seluruh proses tersebut.



# The Problem Without Pipelines

Tanpa pipeline:

- Mudah terjadi data leakage
- Kode sulit dipelihara
- Sulit melakukan tuning
- Rentan kesalahan preprocessing



# Parameter Selection with Preprocessing

Preprocessing harus dilakukan hanya menggunakan data training.

Kesalahan umum:

Melakukan fit scaler pada seluruh dataset sebelum train-test split.

Hal ini menyebabkan data leakage.


```

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(
    cancer.data,
    cancer.target,
    random_state=0
)

scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```


### Penjelasan

fit()

Menghitung parameter preprocessing.

transform()

Mengubah data menggunakan parameter tersebut.

Data test tidak boleh digunakan saat fit.



# Building Pipelines

Pipeline menghubungkan preprocessing dan model menjadi satu objek.


```

from sklearn.pipeline import Pipeline
from sklearn.svm import SVC

pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("svc", SVC())
])

pipe.fit(X_train, y_train)

print(pipe.score(X_test, y_test))
```


### Keuntungan Pipeline

- Lebih rapi
- Mengurangi kesalahan
- Mendukung Grid Search
- Reproducible



# Understanding Pipeline Steps

Pipeline terdiri dari:

1. Transformer
2. Estimator

Transformer:
- StandardScaler
- MinMaxScaler
- PCA

Estimator:
- SVM
- Logistic Regression
- Random Forest



# Using Pipelines in Grid Search

Pipeline dapat langsung digunakan pada Grid Search.


```

from sklearn.model_selection import GridSearchCV

param_grid = {
    "svc__C":[0.1,1,10],
    "svc__gamma":[0.001,0.01,0.1]
}

grid = GridSearchCV(
    pipe,
    param_grid,
    cv=5
)

grid.fit(X_train,y_train)

print(grid.best_params_)
print(grid.best_score_)
```


### Penjelasan Penting

Format:

nama_step__parameter

Contoh:

svc__C

artinya parameter C milik model SVC.



# The General Pipeline Interface

Pipeline bertindak seperti model biasa.

Mendukung:

- fit()
- predict()
- score()
- transform()


```

predictions = pipe.predict(X_test)

print(predictions[:10])
```


# Convenient Pipeline Creation with make_pipeline

Scikit-Learn menyediakan make_pipeline.


```

from sklearn.pipeline import make_pipeline

pipe2 = make_pipeline(
    StandardScaler(),
    SVC()
)

pipe2.fit(X_train,y_train)

print(pipe2.score(X_test,y_test))
```


### Perbedaan

Pipeline()

Nama step ditentukan manual.

make_pipeline()

Nama step dibuat otomatis.



# Accessing Step Attributes

Kita dapat mengakses masing-masing komponen pipeline.


```

pipe.fit(X_train,y_train)

print(pipe.named_steps["svc"])
print(pipe.named_steps["scaler"])
```


### Kegunaan

Memungkinkan inspeksi parameter model setelah training.



# Accessing Attributes in a Grid-Searched Pipeline

Setelah Grid Search selesai:

Model terbaik dapat diakses menggunakan best_estimator_.


```

best_model = grid.best_estimator_

print(best_model)
```


### Penjelasan

best_estimator_

Menyimpan pipeline dengan kombinasi parameter terbaik.



# Grid Searching Preprocessing Steps and Model Parameters

Tidak hanya model.

Preprocessing juga dapat dituning.


```

from sklearn.decomposition import PCA

pipe_pca = Pipeline([
    ("scaler", StandardScaler()),
    ("pca", PCA()),
    ("svc", SVC())
])

param_grid = {
    "pca__n_components":[2,5,10,20],
    "svc__C":[0.1,1,10]
}

grid_pca = GridSearchCV(
    pipe_pca,
    param_grid,
    cv=5
)
```


### Penjelasan

Sekarang Grid Search mencari:

- Jumlah komponen PCA terbaik
- Nilai C terbaik

secara bersamaan.
