# Bag of Words (BoW)

Bag of Words merepresentasikan dokumen berdasarkan frekuensi kemunculan kata.

Urutan kata diabaikan.

Hanya jumlah kemunculan kata yang digunakan.


```

from sklearn.feature_extraction.text import CountVectorizer

texts = [
    "I love machine learning",
    "Machine learning is fun",
    "I love Python"
]

vect = CountVectorizer()

X = vect.fit_transform(texts)

print(vect.get_feature_names_out())
print(X.toarray())
```


### Penjelasan Kode

CountVectorizer()

Membangun vocabulary dari seluruh dokumen.

fit_transform()

Mempelajari vocabulary dan mengubah teks menjadi matriks numerik.

toarray()

Menampilkan matriks dalam bentuk yang mudah dibaca.



# Vocabulary

Vocabulary adalah kumpulan seluruh kata unik yang ditemukan pada dataset.

Contoh:

['fun','is','learning','love','machine','python']



# Sparse Matrix

Data teks biasanya menghasilkan matriks yang sangat besar dan sebagian besar bernilai nol.

Karena itu Scikit-Learn menyimpan hasil vectorization menggunakan sparse matrix.


```

print(type(X))
print(X.shape)
```


# Tokenization

Tokenization adalah proses memecah teks menjadi unit-unit kecil (token).

Contoh:

'I love machine learning'

→

['I','love','machine','learning']



# Text Classification Example

Dataset yang sering digunakan:

IMDb Reviews

Tujuan:

Memprediksi apakah review bersifat positif atau negatif.


```

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(
    X,
    [1,1,0],
    random_state=42
)

model = LogisticRegression()

model.fit(X_train, y_train)

print(model.score(X_test, y_test))
```


### Penjelasan

Logistic Regression merupakan salah satu model paling populer untuk text classification.



# TF-IDF Representation

TF-IDF:

Term Frequency – Inverse Document Frequency

Memberikan bobot lebih tinggi pada kata yang informatif.


```

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()

X_tfidf = tfidf.fit_transform(texts)

print(X_tfidf.shape)
```


### Mengapa TF-IDF?

Kata yang muncul di hampir semua dokumen biasanya kurang informatif.

TF-IDF mengurangi pengaruh kata-kata tersebut.



# Inspecting TF-IDF Features


```

print(tfidf.get_feature_names_out())
```


Setiap kata memperoleh bobot berdasarkan kepentingannya terhadap dokumen.



# Stop Words

Stop words adalah kata yang sering muncul tetapi kurang informatif.

Contoh:

- the
- is
- and
- of


```

vect = CountVectorizer(stop_words="english")

X_stop = vect.fit_transform(texts)

print(vect.get_feature_names_out())
```


### Manfaat Stop Words Removal

- Mengurangi dimensi fitur
- Mengurangi noise
- Mempercepat training



# N-Grams

N-Gram mempertimbangkan urutan kata.

Contoh:

Bigram:

'machine learning'

Trigram:

'introduction to machine'


```

vect = CountVectorizer(
    ngram_range=(1,2)
)

X_ngram = vect.fit_transform(texts)

print(vect.get_feature_names_out())
```


### Penjelasan

ngram_range=(1,2)

Menggunakan:

- Unigram
- Bigram



# Feature Importance in Text Classification


```

vect = TfidfVectorizer()
X_tfidf = vect.fit_transform(texts)

print(X_tfidf.toarray())
```


Model sering memanfaatkan bobot kata untuk menentukan sentimen atau kategori dokumen.



# Combining Text Processing and Models with Pipelines


```

from sklearn.pipeline import make_pipeline

pipe = make_pipeline(
    TfidfVectorizer(),
    LogisticRegression()
)

print(pipe)
```


Pipeline memungkinkan seluruh proses:

Text
↓
Vectorization
↓
Classification

dijalankan secara otomatis.
