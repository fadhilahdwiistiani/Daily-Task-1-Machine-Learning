# 3. Preprocessing and Scaling

Mengapa preprocessing penting?

Banyak algoritma Machine Learning sensitif terhadap skala fitur.

Contoh:

Umur: 18–70

Pendapatan: 1.000.000–100.000.000

Tanpa scaling, fitur dengan nilai besar akan mendominasi model.



# StandardScaler

StandardScaler mengubah data menjadi:

Mean = 0

Standard Deviation = 1


```

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(
    cancer.data,
    cancer.target,
    random_state=1
)

scaler = StandardScaler()

scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(X_train_scaled.mean())
```


### Penjelasan Kode

fit()

Menghitung mean dan standar deviasi data training.

transform()

Mengubah data menggunakan parameter yang diperoleh saat fit.

Penting:

Jangan melakukan fit pada data test.



# 4. Principal Component Analysis (PCA)

PCA adalah teknik dimensionality reduction.

Tujuan:

- Mengurangi jumlah fitur
- Mengurangi noise
- Visualisasi data
- Mempertahankan informasi utama


```

from sklearn.decomposition import PCA

pca = PCA(n_components=2)

X_pca = pca.fit_transform(X_train_scaled)

print(X_train.shape)
print(X_pca.shape)
```


### Penjelasan

Misalnya:

30 fitur → 2 principal components

PCA mencari arah dengan variansi terbesar.

Konsep penting:

- Eigenvector
- Eigenvalue
- Explained Variance



# Explained Variance


```

print(pca.explained_variance_ratio_)
```


Nilai ini menunjukkan berapa banyak informasi yang dipertahankan oleh masing-masing principal component.



# 5. Non-Negative Matrix Factorization (NMF)

NMF merupakan teknik feature extraction.

Perbedaan utama dengan PCA:

PCA:
- Menggunakan komponen positif dan negatif

NMF:
- Hanya nilai non-negatif
- Lebih mudah diinterpretasikan


```

from sklearn.decomposition import NMF

nmf = NMF(
    n_components=10,
    random_state=0,
    max_iter=1000
)

X_nmf = nmf.fit_transform(X_train_scaled - X_train_scaled.min())

print(X_nmf.shape)
```


### Kegunaan NMF

- Topic Modeling
- Image Analysis
- Feature Extraction



# 6. Manifold Learning with t-SNE

t-SNE digunakan untuk visualisasi data berdimensi tinggi.

Tujuan:

Membuat representasi 2D atau 3D yang mudah divisualisasikan.


```

from sklearn.manifold import TSNE

tsne = TSNE(
    random_state=42,
    init="pca"
)

X_tsne = tsne.fit_transform(X_train_scaled)

print(X_tsne.shape)
```


### Penjelasan

t-SNE tidak digunakan untuk training model.

Digunakan untuk:
- Visualisasi
- Eksplorasi dataset
- Analisis cluster



# 7. Clustering

Clustering mengelompokkan data berdasarkan kemiripan.

Tujuan:

Menemukan kelompok alami pada data tanpa label.



# 8. K-Means Clustering

K-Means adalah algoritma clustering paling populer.


```

from sklearn.cluster import KMeans

kmeans = KMeans(
    n_clusters=3,
    random_state=0
)

kmeans.fit(X_train_scaled)

print(kmeans.labels_[:10])
```


### Cara Kerja K-Means

1. Pilih centroid awal
2. Assign data ke centroid terdekat
3. Hitung centroid baru
4. Ulangi hingga konvergen



# 9. Agglomerative Clustering

Metode hierarchical clustering.


```

from sklearn.cluster import AgglomerativeClustering

agg = AgglomerativeClustering(
    n_clusters=3
)

labels = agg.fit_predict(X_train_scaled)

print(labels[:10])
```


### Penjelasan

Setiap data awalnya merupakan cluster tersendiri.

Kemudian cluster digabung secara bertahap.



# 10. DBSCAN

Density-Based Spatial Clustering of Applications with Noise.


```

from sklearn.cluster import DBSCAN

dbscan = DBSCAN(
    eps=2,
    min_samples=5
)

clusters = dbscan.fit_predict(X_train_scaled)

print(clusters[:10])
```


### Kelebihan DBSCAN

- Tidak perlu menentukan jumlah cluster
- Dapat mendeteksi outlier
- Cocok untuk bentuk cluster kompleks
